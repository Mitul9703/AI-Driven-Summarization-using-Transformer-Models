```python
import re
from pdfminer.high_level import extract_text
import pandas as pd
import os
import torch
from transformers import T5ForConditionalGeneration, T5Tokenizer
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize
```

```python
model_save_path = '/content/drive/MyDrive/AI_WEBAPP/t5_finetuned_model_new_target_sum_100__small.bin'
tokenizer = T5Tokenizer.from_pretrained('t5-small')
model = T5ForConditionalGeneration.from_pretrained('t5-small')
model.load_state_dict(torch.load(model_save_path))
model.eval()
```

```python



def extract_text_pdfminer(pdf_path):
    text = extract_text(pdf_path)
    return remove_header_footer(text)

def remove_header_footer(text):

    pattern = r'^Proceedings of .*$'

    pattern2 = r"(Authorized licensed use limited to: Sri Sivasubramanya Nadar College of Engineering\. Downloaded on November 17,2023 at 06:03:49 UTC from IEEE Xplore\.  Restrictions apply\. )|(IEEE Xplore Part Number: [A-Z0-9-]+; ISBN: )?\b978-\d{1,5}-\d{1,7}-\d{1,6}-\d{1,2}\b(\/22\/\$\d{1,}\.00 ©\d{4} IEEE)?"
    #pattern3 = "Authorized licensed use limited to: Sri Sivasubramanya Nadar College of Engineering\. Downloaded on November 17,2023 at 06:03:49 UTC from IEEE Xplore\.  Restrictions apply\. "

    text_to_remove = "Authorized licensed use limited to: Sri Sivasubramanya Nadar College of Engineering. Downloaded on November 13,2023 at 16:18:00 UTC from IEEE Xplore.  Restrictions apply."
    sentences = re.split('(?<=[.!?]) +', text)

    sentences = [sentence for sentence in sentences if "Sri Sivasubramanya Nadar College of Engineering" not in sentence and "Downloaded on" not in sentence]
    text = " ".join(sentences)
    cleaned_text = re.sub(pattern2, '', text)
    cleaned_text = cleaned_text.replace(text_to_remove, '')
    cleaned_text = re.sub(r'\b\d+\b', '', cleaned_text)

    lines = cleaned_text.split('\n')
    cleaned_lines = [line for line in lines if not re.match(pattern, line.strip())]


    cleaned_text = '\n'.join(cleaned_lines)

    return cleaned_text







def extract_key_sections(text):

    methodology_pattern = r'METHODOLOGY(.*?)(?=RESULTS|DISCUSSION|CONCLUSION)'
    results_pattern = r'RESULTS(.*?)(?=DISCUSSION|CONCLUSION)'
    conclusion_pattern = r'CONCLUSION(.*?)REFERENCES'


    methodology = re.search(methodology_pattern, text, re.DOTALL | re.IGNORECASE)
    results = re.search(results_pattern, text, re.DOTALL | re.IGNORECASE)
    conclusion = re.search(conclusion_pattern, text, re.DOTALL | re.IGNORECASE)


    methodology_text = methodology.group(1).strip() if methodology else ''
    results_text = results.group(1).strip() if results else ''
    conclusion_text = conclusion.group(1).strip() if conclusion else ''

    return {'Methodology': methodology_text, 'Results': results_text, 'Conclusion': conclusion_text}


def extract_relevant_sections(text):
    # Define patterns for the start of the body and conclusion sections
    body_start_pattern = r'II\.'
    conclusion_pattern = r'CONCLUSION(.*?)REFERENCES'

    # Extracting Body and Conclusion sections
    body_start_match = re.search(body_start_pattern, text, re.IGNORECASE)
    body_start_index = body_start_match.start() if body_start_match else None
    conclusion_match = re.search(conclusion_pattern, text, re.DOTALL | re.IGNORECASE)
    body_text = text[body_start_index:conclusion_match.start()] if body_start_index and conclusion_match else ''
    conclusion_text = conclusion_match.group(1).strip() if conclusion_match else ''

    results_pattern = r'RESULTS(.*?)(?=DISCUSSION|CONCLUSION)'
    irrelevant_sections = ['BACKGROUND', 'RELATED WORK']
    irrelevant_section_pattern = '|'.join([f'{i}\\.' for i in irrelevant_sections])

    # results_match = re.search(results_pattern, body_text, re.DOTALL | re.IGNORECASE)
    # results_text = results_match.group(1).strip() if results_match else ''
    for pattern in irrelevant_sections:
        body_text = re.sub(pattern + r'.*?(?=II\.|III\.|IV\.|V\.|VI\.|VII\.|VIII\.|IX\.|X\.|XI\.|XII\.|XIII\.|XIV\.|XV\.|CONCLUSION)', '', body_text, flags=re.DOTALL | re.IGNORECASE)

    return {'Body': body_text, 'Conclusion': conclusion_text}



def extract_body_and_conclusion(text):

    body_start_pattern = r'II\.'
    conclusion_pattern = r'CONCLUSION(.*?)REFERENCES'

    body_start_match = re.search(body_start_pattern, text, re.IGNORECASE)
    body_start_index = body_start_match.start() if body_start_match else None

    conclusion_match = re.search(conclusion_pattern, text, re.DOTALL | re.IGNORECASE)
    conclusion_text = conclusion_match.group(1).strip() if conclusion_match else ''


    body_text = text[body_start_index:conclusion_match.start()] if body_start_index and conclusion_match else ''

    return {'Body': body_text, 'Conclusion': conclusion_text}




from io import BytesIO

def process_paper(pdf_file):

    file_stream = BytesIO(pdf_file.read())
    extracted_text = extract_text_pdfminer(file_stream)
    return extract_relevant_sections(extracted_text)

def clean_summary(summary):
    def capitalize_first_letter(sentence):
        return sentence.strip()[0].upper() + sentence.strip()[1:] if sentence else sentence

    cleaned_summary = re.sub(r'\[\]', '', summary)
    cleaned_summary = re.sub(r'\[\s*\]', '', cleaned_summary)
    cleaned_summary = re.sub(r'\(\s*\)', '', cleaned_summary)
    sentences = re.split('(?<=[.!?]) +', cleaned_summary)
    capitalized_sentences = [capitalize_first_letter(sentence) for sentence in sentences]
    capitalized_text = ' '.join(capitalized_sentences)

    return capitalized_text



def split_into_sentences(text, max_length=500):
    sentences = sent_tokenize(text)
    chunks = []
    current_chunk = ""
    for sentence in sentences:
        if len(current_chunk) + len(sentence) <= max_length:
            current_chunk += ' ' + sentence
        else:
            chunks.append(current_chunk.strip())
            current_chunk = sentence
    if current_chunk:
        chunks.append(current_chunk.strip())
    return chunks

def generate_summary_for_chunk(text_chunk):
    input_ids = tokenizer.encode("summarize: " + text_chunk, return_tensors="pt", max_length=512, truncation=True)
    summary_ids = model.generate(input_ids, max_length=150, length_penalty=2.0, num_beams=4, early_stopping=True)

    if summary_ids.numel() == 0:
        return ""

    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)


def run_summary(pdf) :

    extracted_sections = process_paper(pdf)
    chunks = split_into_sentences(extracted_sections['Body'] + extracted_sections['Conclusion'])
    summaries = [generate_summary_for_chunk(chunk) for chunk in chunks]
    combined_summary = ' '.join(filter(None, summaries))
    cleaned_combined_summary = clean_summary(combined_summary)
    print("Cleaned Summary:", cleaned_combined_summary)

    return cleaned_combined_summary


```

```python
from flask import Flask, render_template, request, redirect
from flask_ngrok import run_with_ngrok
```

```python
app = Flask(__name__, static_folder="static", template_folder="templates")

run_with_ngrok(app)

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/upload', methods=['POST'])
def upload():
    if 'pdf_file' not in request.files:
        return redirect(request.url)
    pdf_file = request.files['pdf_file']
    print(pdf_file)
    if pdf_file.filename == '':
        return redirect(request.url)

    if pdf_file:
        # pdf_path = os.path.join('/tmp', pdf_file.filename)
        # pdf_file.save(pdf_path)

        text = run_summary(pdf_file)

        return render_template('index.html', pdf_file = pdf_file, text=text)


app.run()
```
